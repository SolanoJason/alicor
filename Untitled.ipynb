```python
values = train.groupby(['customer_id', 'product_id']).agg(cantidad_venta=('cantidad_venta', 'sum'))
values
```

```python
matrix = pd.pivot_table(values, index='customer_id', columns='product_id', values='cantidad_venta')
matrix.fillna(0, inplace=True)
matrix
```

```python
customer_similarity = np.corrcoef(matrix.values)
customer_similarity = pd.DataFrame(customer_similarity, index=matrix.index, columns=matrix.index)
customer_similarity.head()
```

```python
def get_products(picked_customer_id):
    customer_similarity_copy = customer_similarity.copy()
    customer_similarity_copy.drop(picked_customer_id, axis=0, inplace=True)
    n = 50
    similar_customers = customer_similarity_copy[picked_customer_id].sort_values(ascending=False)[:50]
    similar_customers_products = matrix.loc[similar_customers.index].dropna(axis=1, how='all')
    weighted_similar_customers_products = similar_customers_products.mul(similar_customers, axis=0)
    recommendend_products = weighted_similar_customers_products.sum(axis=0).sort_values(ascending=False)[:n]
    return recommendend_products

def get_user_products(uid, dataset):
    """
    Retornar los productos que un usuario ha comprado en el dataset train o test
    """
    comprados = dataset[dataset.customer_id == uid].product_id.unique()
    return comprados
```
