```python
df = pd.merge(ratings, movies, how='left', on='movieId')
df.head()
```

```python
grouped = df.groupby('title')
agg_ratings = grouped.agg(mean_rating = pd.NamedAgg('rating','mean'), number_of_ratings = pd.NamedAgg('rating', 'count'))
agg_ratings.head()
```

```python
agg_ratings_GT100 = agg_ratings[agg_ratings.number_of_ratings > 100]
agg_ratings_GT100.reset_index(inplace=True)
agg_ratings_GT100.info()
```

```python
agg_ratings_GT100.sort_values('mean_rating', ascending=False)
```

```python
sns.jointplot(agg_ratings_GT100, x='mean_rating', y='number_of_ratings')
```

```python
df_GT100 = df[df.title.isin(agg_ratings_GT100.title)]
df_GT100.head()
```

```python
print(f'There are {df_GT100.userId.nunique()} unique users', f'There are {df_GT100.movieId.nunique()} unique movies', f'There are {df_GT100.rating.nunique()} unique ratings', sep='\n')
print(f'The unique ratings are: {sorted(df_GT100.rating.unique())}')
```

```python
matrix = df_GT100.pivot_table(index='userId', columns='title', values='rating')
matrix.head()
```

```python
matrix_norm = matrix.subtract(matrix.mean(axis=1), axis=0)
matrix_norm.head()
```

```python
user_similarity = matrix_norm.T.corr()
user_similarity.head()
```

```python
picked_userid = 152
user_similarity.drop(index=picked_userid, inplace=True)
user_similarity.head()
```

```python
n = 10
user_similarity_threshold = 0.3
similar_users = user_similarity.loc[user_similarity[picked_userid] >= user_similarity_threshold, picked_userid].sort_values(ascending=False)[:10]
print(f'The similar users for user {picked_userid} are: {similar_users}')
```

```python
picked_userid_watched = matrix_norm.loc[picked_userid].dropna()
picked_userid_watched.head()
```

```python
similar_user_movies = matrix_norm.loc[similar_users.index].dropna(axis=1, how='all')
similar_user_movies.head()
```

```python
similar_user_movies.drop(labels=picked_userid_watched.index, axis=1, errors='ignore', inplace=True)
similar_user_movies.head()
```

```python
weighted_similar_user_movies = similar_user_movies.mul(similar_users, axis=0).copy()
weighted_similar_user_movies
```

```python
recommendend_movies = weighted_similar_user_movies.apply(lambda c: similar_users[similar_users.index.isin(c[c.notna()].index)].sum()).sort_values(ascending=False)
recommendend_movies.head()
```
